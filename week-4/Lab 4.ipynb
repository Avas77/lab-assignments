### Question 1: Read the content of the `abcnews-date-text.csv` and perform the following:
<hr style="border:1px solid orange"> </hr>

> **Q1.1.** Preprocess the `headline_text` field content by performing the necessary steps to make it ready for applying topic modeling algorithms. Store the resulted preprocessed text into a list called `headlines`

> **Q1.2.** Using `Explicit Symantic Analysis` algorithm, find and print the top `20` related documents to the following document: 
>>   \['price', 'sugar', 'sweet'\]. 

> **Q1.3.** Using only the first `50` headlines, perform a document clustering using `AgglomerativeClustering` algorithm and plot the resulted dendrogram.

> **Q1.4.** Using `Latent Dirichlett Allocation` algorithm, perform topic modeling to find `20` topics from the headlines. Then, print the top `10` words for each topic. 

> **Q1.5.** Using `Non-Negative Matrix Factorization` algorithm, perform topic modeling to find `20` topics from the headlines. Then, print the top `10` words for each topic. 

```python
import string
import pandas as pd
import nltk
from nltk.corpus import stopwords
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt

news_df = pd.read_csv("abcnews-date-text.csv")
news_df_short = news_df[:100]
```

```python
# Q1.1. 
def preprocess(text):
    news = []
    lower_text = text.lower()
    tokens = text.split()
    stop_words = set(stopwords.words("english"))
    for token in tokens:
        if token not in stop_words and token.isalpha():
            news.append(token)
    return news
```

```python
headlines = news_df_short["headline_text"].apply(preprocess).tolist()
headlines
```

```python
# Q1.2. 
from nltk.text import TextCollection

corpus = TextCollection(headlines)
corpus
```

```python
TFIDF = []

for word in headlines:
    score = {}
    for token in set(word):
        score[token] = corpus.tf_idf(token, word)

    TFIDF.append(score)
    
```

```python
TFIDF
```

```python
word_list = list(corpus.vocab().keys())
word_dict = dict(zip(word_list, np.arange(len(word_list))))
```

```python
word_dict
```

```python
def term_document_matrix(TFIDF, word_list, word_dict):
    vocabulary_size = len(word_dict)
    number_documents = len(TFIDF)
    TDM = np.zeros((vocabulary_size, number_documents))
    for doc in range(number_documents):
        document = TFIDF[doc]
        for word in document.keys():
            pos = word_dict[word]
            TDM[pos, doc] = document[word]
    return TDM
```

```python
TDM = term_document_matrix(TFIDF, word_list, word_dict)
TDM.shape
```

```python
new_docs = ['offer', 'police', 'station']
```

```python
def find_related_docs(docs, TDM):
    new_vector = np.zeros(TDM.shape[1]) 
    
    for word in docs:
        pos = word_dict[word]
        new_vector += TDM[pos, :]
        doc_list = sorted(zip(range(TDM.shape[1]), new_vector),
                         key=lambda x:x[1], reverse=True)
        return doc_list
```

```python
related = find_related_docs(new_docs, TDM)
```

```python
for tweet, score in related[:10]:
    print(score, tweet, " ".join(headlines[tweet]))
```

```python
# Q1.3. 
from sklearn.cluster import AgglomerativeClustering 
from scipy.cluster.hierarchy import dendrogram 

def plot_dendrogram(model, **kwargs):
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1
            else:
                current_count += counts[child_idx - n_samples]
            counts[i] = current_count
    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)
    dendrogram(linkage_matrix, **kwargs)
            

```

```python
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)
model = model.fit(TDM.T[:50])
```

```python
model.n_features_in_
```

```python
plt.figure(figsize=(16,6))
plot_dendrogram(model)
```

```python
# Q1.4. 
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components=20, random_state= 42)
lda.fit(TDM.T)
```

```python
lda.components_.shape
```

```python
for i, topic in enumerate(lda.components_):
    print("Topic {}: {}".format(i + 1, ", ".join([str(x) for x in np.array(word_list)[topic.argsort()[-10:]]])))
```

```python
# Q1.5. 
from sklearn.decomposition import NMF 

nmf = NMF(n_components=10, init="nndsvd", max_iter=2000)
W = nmf.fit_transform(TDM.T)
H = nmf.components_
err = nmf.reconstruction_err_
```

```python
W.shape
```

```python
for i, topic in enumerate(H):
    print("Topic {}: {}".format(i + 1, ", ".join([str(x) for x in np.array(word_list)[topic.argsort()[-10:]]])))
```
