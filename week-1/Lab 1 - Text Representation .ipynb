```python
import nltk

file = open("ShakespearesSonnets.txt", encoding="utf-8")
sonnets = file.read()
sonnets
```

```python
import string

def tokenize(text):
    punctuation = set(string.punctuation)
    text_words = []
    for word in nltk.tokenize.WordPunctTokenizer().tokenize(text):
        if word in punctuation or not word.isalpha():
            continue
        
        text_words.append(word)
        
    return text_words
```

```python
text_tokens = tokenize(sonnets)
text_tokens
```

```python
sentences = nltk.sent_tokenize(sonnets, "english")
sentences
```

```python
import pandas as pd

one_hot = pd.get_dummies(text_tokens)
one_hot
```

```python
temp = one_hot.astype("str")
temp[temp=="0"] = ""
temp = pd.DataFrame(temp)
temp
```

```python
import numpy as np

word_dict = dict(zip(one_hot.columns, np.arange(one_hot.shape[1])))
word_dict
```

```python

pd.DataFrame(one_hot.sum(), columns=["Count"])
```

```python
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
stop_words
```

```python
filtered_tokens = [tokens for tokens in text_tokens if tokens not in stop_words]
filtered_tokens
```

```python
corpus = nltk.sent_tokenize(sonnets)
corpus
```

```python
corpus = [tokenize(doc) for doc in corpus]
corpus
```

```python
from nltk.text import TextCollection 

nlp = TextCollection(corpus)
```

```python
TFIDF = []

for doc in corpus:
    current = {}
    for token in doc:
        current[token] = nlp.tf_idf(token, doc)
    TFIDF.append(current)

TFIDF
```

```python
from nltk.util import ngrams

def get_ngrams(text, length):
    n_grams = ngrams(tokenize(text), length)
    return [" ".join(grams) for grams in n_grams]
```

```python
get_ngrams(sonnets, 4)
```

```python
from nltk.collocations import BigramCollocationFinder 
from nltk.metrics.association import BigramAssocMeasures

bigrams = BigramCollocationFinder.from_words(text_tokens)
scored = bigrams.score_ngrams(BigramAssocMeasures.likelihood_ratio)
scored
```
